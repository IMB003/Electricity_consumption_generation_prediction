```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("non_renewable_data.csv")
#selecting the columns with value western
df = df[df['Region'] == 'Western']
df.columns
```

```python
# Create input and output features
# what is feature_name?
# what is X_data? 
# what is y_data?
# y_data = df['Thermal Generation Actual (in MU)',]
# remove 1st and 2nd column from feature_names
#selecting columns with value 'Thermal Generation Actual (in MU)' and 'Nuclear Generation Actual (in MU)' and 'Hydro Generation Actual (in MU)'
df = df[['Date','Thermal Generation Actual (in MU)','Nuclear Generation Actual (in MU)','Hydro Generation Actual (in MU)']]
df
```

```python
# convert string to float
# remove comma from the string except date column
df['Thermal Generation Actual (in MU)'] = df['Thermal Generation Actual (in MU)'].str.replace(',', '').astype(float)
df
```

```python
#checking for outliers and plotting boxplot of width 10 and height 5
plt.figure(figsize=(10,10))
sns.boxplot(data=df)
plt.show()
```

```python
#removing outliers
#removing outliers from 'Thermal Generation Actual (in MU)'
Q1 = df['Thermal Generation Actual (in MU)'].quantile(0.25)
Q3 = df['Thermal Generation Actual (in MU)'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Thermal Generation Actual (in MU)'] < (Q1 - 1.5 * IQR)) |(df['Thermal Generation Actual (in MU)'] > (Q3 + 1.5 * IQR)))]
df
```

```python
#removing outliers from 'Nuclear Generation Actual (in MU)'
Q1 = df['Nuclear Generation Actual (in MU)'].quantile(0.25)
Q3 = df['Nuclear Generation Actual (in MU)'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Nuclear Generation Actual (in MU)'] < (Q1 - 1.5 * IQR)) |(df['Nuclear Generation Actual (in MU)'] > (Q3 + 1.5 * IQR)))]
df
```

```python
#removing outliers from 'Hydro Generation Actual (in MU)'
Q1 = df['Hydro Generation Actual (in MU)'].quantile(0.25)
Q3 = df['Hydro Generation Actual (in MU)'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Hydro Generation Actual (in MU)'] < (Q1 - 1.5 * IQR)) |(df['Hydro Generation Actual (in MU)'] > (Q3 + 1.5 * IQR)))]
df
```

```python
#plotting boxplot after removing outliers
plt.figure(figsize=(10,10))
sns.boxplot(data=df)
plt.show()
```

```python
# plotting graph of Date vs 'Thermal Generation Actual (in MU)'
plt.figure(figsize=(20,10))
plt.plot(df['Date'],df['Thermal Generation Actual (in MU)'])
plt.xlabel('Date')
plt.ylabel('Thermal Generation Actual (in MU)')
plt.show()
```

```python
# plotting graph of Date vs 'Nuclear Generation Actual (in MU)'
plt.figure(figsize=(20, 10))
plt.plot(df['Date'], df['Nuclear Generation Actual (in MU)'])
plt.xlabel('Date')
plt.ylabel('Nuclear Generation Actual (in MU)')
plt.show()
```

```python
#using DecisionTreeRegressor to predict 'Thermal Generation Actual (in MU)'
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#splitting data into train and test of thermal generation
X_train, X_test, y_train, y_test = train_test_split(df['Date'], df['Thermal Generation Actual (in MU)'], test_size=0.2, random_state=42)

#converting date into integer
X_train = pd.to_datetime(X_train)
X_train = X_train.map(pd.Timestamp.to_julian_date)
X_train = X_train.values.reshape(-1,1)

X_test = pd.to_datetime(X_test)
X_test = X_test.map(pd.Timestamp.to_julian_date)
X_test = X_test.values.reshape(-1,1)

#fitting the model
model = DecisionTreeRegressor()
model.fit(X_train,y_train)

#predicting the values
y_pred = model.predict(X_test)

#calculating the mean squared error
mse = mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)

#calculating the r2 score
r2 = r2_score(y_test,y_pred)
print("R2 Score:",r2)

#accuracy of the model
print("Accuracy of the model:",model.score(X_test,y_test))
```
